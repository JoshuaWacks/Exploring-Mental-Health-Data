Look at IterativeImputer

```python
def fill_in_missing_values(df,numerical_cols,categorical_cols):
    imputer = SimpleImputer(missing_values=np.nan,strategy='median',add_indicator=True)

    for col in numerical_cols:
        if sum(pd.isnull(df[col])) > 0:
            df[col] = imputer.fit_transform(df[[col]])
    
    return df
```

```python
numeric_cols = org_student_data.columns.to_list()
numeric_cols.remove('id')
numeric_cols.remove('Name')
numeric_cols.remove('Gender')
numeric_cols.remove('City')
numeric_cols.remove('Sleep Duration')
numeric_cols.remove('Dietary Habits')
numeric_cols.remove('Degree')
numeric_cols.remove('Family History of Mental Illness')
numeric_cols.remove('Have you ever had suicidal thoughts ?')
numeric_cols.remove('Depression')
```

```python
temp = fill_in_missing_values(org_student_data,numeric_cols,categorical_cols=['name'])
```
