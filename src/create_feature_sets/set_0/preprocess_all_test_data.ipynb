```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

from test_prof_preprocess import TestPreprocess
```

```python
test_student_data =pd.read_csv('../../data/data_split_0/student_test_data.csv')
```

```python
cols_to_drop=['Name']

imputer = SimpleImputer(missing_values=np.nan,strategy='median')


student_preprocessor = TestPreprocess(df=test_student_data,
                                             cols_to_drop=cols_to_drop,
                                             imputer=imputer,
                                             cat_encoding_type="frequency",
                                             is_student=True)
```

```python
student_preprocessor.preprocess()
```

```python
student_clean_df=student_preprocessor.df
```

```python
student_clean_df[student_clean_df.isnull().any(axis=1)]
```

```python
student_clean_df
```

```python
student_clean_df.to_csv('../../data/clean_data_0/student_test_data.csv',index=False)
```
